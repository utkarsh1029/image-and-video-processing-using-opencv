Contours are continuous curves or boundaries that represent the shape or outline of an object in an image.

```python
!pip install opencv-python-headless
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import cv2_imshow

# Step 1: Load the image
image = cv2.imread('/content/light.jpg')

# Step 2: Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 3: Apply edge detection (Canny edge detector)
edges = cv2.Canny(gray, threshold1=100, threshold2=200)

# Display the original image and the edges
cv2_imshow(image)  # Use cv2_imshow instead of cv2.imshow
cv2_imshow(edges)  # Use cv2_imshow instead of cv2.imshow


# Step 4: Find contours
contours, hierarchy = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Step 5: Draw contours on the original image
output_image = image.copy()
cv2.drawContours(output_image, contours, -1, (0, 255, 0), 2)

# Display the result
cv2_imshow(output_image)  # Use cv2_imshow instead of cv2.imshow

# Save the result (optional)
cv2.imwrite('contours_output.jpg', output_image)

cv2_imshow(output_image)  # Use cv2_imshow instead of cv2.imshow

```


  
threshold1 and threshold2 are the lower and upper thresholds for edge detection.
You can adjust these values depending on the level of sensitivity you need.

cv2.RETR_EXTERNAL finds only external contours.
    
cv2.CHAIN_APPROX_SIMPLE simplifies the contour and removes redundant points.

The contours are drawn in green ((0, 255, 0)) with a thickness of 2 pixels.
You can change the color and thickness to suit your needs.
