```python
from PIL import Image
# Open a grayscale image
image = Image.open("/content/light.jpg").convert("L")  # Convert to grayscale

# Convert image to binary by applying a threshold
threshold = 128  # Pixel values above this will become white (255), below will become black (0)
binary_image = image.point(lambda p: 255 if p > threshold else 0)

# Show the binary image
display(binary_image)

# Save the binary image
binary_image.save("binary_image.png")
```

```python
from PIL import Image
import numpy as np

# Load the multispectral image (let's assume it has 4 bands)
image = Image.open("/content/download.jpg")

# Convert the image into a NumPy array
image_array = np.array(image)

# Display the shape of the array
print(image_array.shape)

# Extract individual bands (assume the image has 4 bands)
# In this case, we're assuming the bands are ordered as:
# Band 0 - Red, Band 1 - Green, Band 2 - Blue, Band 3 - Near-Infrared (NIR)
red_band = image_array[:, :, 0]  # Red
#green_band = image_array[:, :, 1]  # Green
blue_band = image_array[:, :, 2]  # Blue


# Create an RGB image using the red, green, and blue bands
#rgb_image = np.stack((red_band, green_band, blue_band), axis=-1)
# Create an RGB image using the red and blue bands
rgb_image = np.stack((red_band, blue_band), axis=-1)

# Convert back to a Pillow image
rgb_image_pil = Image.fromarray(rgb_image)

# Save or display the RGB image
rgb_image_pil.save("converted_rgb_image.png")
display(rgb_image_pil)
```

```python
from PIL import Image, ImageDraw

# Create a blank image with a white background
width, height = 500, 500
image = Image.new("RGB", (width, height), color="white")
draw = ImageDraw.Draw(image)

# Drawing vector-like shapes

# Draw a rectangle
draw.rectangle([50, 50, 200, 200], outline="blue", width=5)

# Draw a circle
draw.ellipse([250, 50, 400, 200], outline="green", width=5)

# Draw a line
draw.line([0, 0, width, height], fill="red", width=5)

# Draw a polygon (triangle)
draw.polygon([(150, 300), (200, 350), (100, 350)], outline="purple", fill="yellow", width=5)

# Save the image
image.save("vector_style_image.png")
display(image)
```

```python
#Resizing an Image
from PIL import Image

# Open an image
img = Image.open("/content/light.jpg")

# Resize the image to 200x300 pixels
resized_img = img.resize((200, 300))
display(resized_img)

# Save the resized image
resized_img.save("resized_example.jpg")
```

```python
#Maintaining Aspect Ratio
#To resize while maintaining the original aspect ratio, use the thumbnail() method:
# Resize with aspect ratio
img.thumbnail((200, 200))
display(img)

# Save the resized image
img.save("thumbnail_example.jpg")
```

```python
#Use the crop() method to extract a portion of the image.
# Define the cropping area (left, upper, right, lower)
crop_area = (50, 50, 200, 200)
cropped_img = img.crop(crop_area)
display(cropped_img)

# Save the cropped image
cropped_img.save("cropped_example.jpg")
```

```python
#The rotate() method rotates the image by a specified angle.
# Rotate the image by 90 degrees
rotated_img = img.rotate(90)
display(rotated_img)

# Save the rotated image
rotated_img.save("rotated_example.jpg")
```

```python
# Flip the image horizontally
flipped_img = img.transpose(Image.FLIP_LEFT_RIGHT)
display(flipped_img)

# Save the flipped image
flipped_img.save("flipped_example.jpg")
# Flip the image vertically
flipped_img_v = img.transpose(Image.FLIP_TOP_BOTTOM)
display(flipped_img_v)

# Save the flipped image
flipped_img_v.save("flipped_example_v.jpg")
```

```python
# Transpose the image (e.g., 90-degree rotation)
transposed_img = img.transpose(Image.ROTATE_90)
display(transposed_img)

# Save the transposed image
transposed_img.save("transposed_example.jpg")
```

```python
import cv2
import numpy as np

# Read the image - Replace 'input_image.jpg' with the actual image file path
image = cv2.imread('/content/light.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if image is None:
    print("Error: Could not open or read the image file.")
else:
    # Create a kernel
    kernel = np.ones((5, 5), np.uint8)

    # Apply erosion
    eroded_image = cv2.erode(image, kernel, iterations=1)

    # Apply dilation
    dilated_image = cv2.dilate(image, kernel, iterations=1)

    # Show the results (using cv2_imshow if in a Colab environment)
    from google.colab.patches import cv2_imshow
    cv2_imshow(eroded_image)
    cv2_imshow(dilated_image)
```
