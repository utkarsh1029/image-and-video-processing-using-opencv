Thresholding: Pixels greater than the threshold are set to 255 (white), and the rest are set to 0 (black). It's a basic form of image segmentation where you can control the threshold value to segment different objects.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('/content/images.jpg', cv2.IMREAD_GRAYSCALE)

# Apply simple thresholding
_, thresholded = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Plot the images
plt.figure(figsize=(10, 7))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(thresholded, cmap='gray')
plt.title('Thresholded Image')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Read the image and convert it to RGB
image = cv2.imread('/content/images.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(image_rgb)
# Reshape the image to 2D array of pixels
pixels = image_rgb.reshape(-1, 3)
print("Reshape the image to 2D array of pixels",pixels)

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(pixels)

# Get the segmented image (replacing each pixel with the centroid color)
segmented_image = kmeans.cluster_centers_[kmeans.labels_]
segmented_image = segmented_image.reshape(image_rgb.shape).astype(int)

# Plot the segmented image
plt.figure(figsize=(10, 7))
plt.imshow(segmented_image)
plt.title('K-Means Segmented Image')
plt.show()
```

```python
# Load the image
image = cv2.imread('/content/images.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Canny edge detection
edges = cv2.Canny(image, 100, 200)

# Plot the original and edge-detected image
plt.figure(figsize=(10, 7))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection (Canny)')
plt.show()
```
